```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import preprocessing
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('data.csv',engine='python')
data
```

```python
y=data['status']
data.drop(['status'],axis=1,inplace=True)#axis=1删除一列
```

### 处理缺失值

```python

missing_data=data.isnull().sum()
for index in missing_data.keys():
    if missing_data[index]==0:
        del missing_data[index]
  

missing_data
```

```python
for index in missing_data.keys():
    if missing_data[index]>2000:
        print('缺失值过多的列',index)
        del missing_data[index]
missing_data
```

```python
data.drop(['student_feature'],axis=1,inplace=True)
```

```python
#对于数值型数据采用平均值补充
for index in missing_data.keys():
    if data[index].dtype!=object:
        data[index].fillna(data[index].mean(),inplace=True)
    else:
        print('object类型有',index,missing_data[index])
```

```python
data.isnull().sum()
```

```python
#剩余没处理的缺失值
data['reg_preference_for_trad'].value_counts()
```

```python
data['reg_preference_for_trad'].fillna('其他城市',inplace=True)
```

```python
#删除id列
data.drop(['id_name'],axis=1,inplace=True)
```

```python
#众数补充
data['latest_query_time'].fillna(data['latest_query_time'].mode()[0],inplace=True)
data['loans_latest_time'].fillna(data['loans_latest_time'].mode()[0],inplace=True)
```

### 对数值型归一化

```python
for i in data.columns:
    if data[i].dtype!=object:
        data[i]=(data[i]-data[i].min())/(data[i].max()-data[i].min())
data.head()
```

### 对非数值类型One_hot处理
  one-hot的基本思想：将离散型特征的每一种取值都看成一种状态，若你的这一特征中有N个不相同的取值，那么我们就可以将该特征抽象成N种不同的状态，one-hot编码保证了每一个取值只会使得一种状态处于“激活态”，也就是说这N种状态中只有一个状态位值为1，其他状态位都是0。

```python
data.drop(['bank_card_no'],axis=1,inplace=True)
```

```python
data.drop(['source'],axis=1,inplace=True)
```

```python
data['time1'].tail()
```

```python
data['year1']=[i.year for i in data['time1']]
```

```python
data['month1']=[i.month for i in data['time1']]
data['day1']=[i.day for i in data['time1']]
data['year2']=[i.year for i in data['time2']]
data['month2']=[i.month for i in data['time2']]
data['day2']=[i.day for i in data['time2']]

```

```python
#数据集划分
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data,y,train_size=0.7, random_state=2018)
```

### 不同模型的应用

```python
from sklearn.linear_model import LogisticRegression#逻辑回归
lr=LogisticRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
svc.score(x_test,y_test)
```

```python

from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier()
DTC.fit(x_train,y_train)
DTC.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=20,random_state = 2019)
forest.fit(x_train,y_train)
forest.score(x_test,y_test)
```

```python
from xgboost import XGBClassifier
xgb_model = XGBClassifier(
 learning_rate =0.05,
 n_estimators=1000,
 max_depth=5,
 min_child_weight=2,
 gamma=0,
 subsample=0.8,
 colsample_bytree=0.8,
 nthread=3,
 seed=2019)
```

```python

xgb_model.fit(x_train,y_train)
xgb_model.score(x_test,y_test)
```

# 模型评价

```python
#引入要用到的评价函数
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_curve,auc
```

```python
def score(x_test,y_test,model):
    y_predict=model.predict(x_test)
    accuracy_=accuracy_score(y_test,y_predict)#准确率
    precision_=precision_score(y_test,y_predict)#精确率
    f1_=f1_score(y_test,y_predict)#F1_score
    recall_=recall_score(y_test,y_predict)#召回率
    auc_=roc_auc_score(y_test,y_predict)#auc值
    
    fpr,tpr,thresholds=roc_curve(y_test,y_predict)#求ROC
    plt.plot(fpr,tpr,'b',label='AUC = %0.4f'% auc_)
    plt.plot([0,1],[0,1],'r--',label= 'Random guess')
    plt.legend(loc='lower right')
    plt.title('ROCcurve')
    plt.xlabel('false positive rate')
    plt.ylabel('true positive rate')
    plt.show()
    
    return accuracy_,precision_,f1_,recall_,auc_
```

```python
models=[forest,lr,svc,DTC,xgb_model]
for model in models:
    accuracy_,precision_,f1_,recall_,auc_=score(x_test,y_test,model)
    print('模型:{}\n准确率:{}\n精确率:{}\nF1值:{}\n召回率:{}\nauc值:{} \n\n '.format(model,
                                                                   accuracy_,
                                                                   precision_,
                                                                   f1_,
                                                                   recall_,
                                                                   auc_))
    
```
