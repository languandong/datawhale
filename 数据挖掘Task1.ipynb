```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import preprocessing
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('data.csv',engine='python')
data
```

```python
y=data['status']
data.drop(['status'],axis=1,inplace=True)#axis=1删除一列
```

### 处理缺失值

```python

missing_data=data.isnull().sum()
for index in missing_data.keys():
    if missing_data[index]==0:
        del missing_data[index]
  

missing_data
```

```python
for index in missing_data.keys():
    if missing_data[index]>2000:
        print('缺失值过多的列',index)
        del missing_data[index]
missing_data
```

```python
data.drop(['student_feature'],axis=1,inplace=True)
```

```python
#对于数值型数据采用平均值补充
for index in missing_data.keys():
    if data[index].dtype!=object:
        data[index].fillna(data[index].mean(),inplace=True)
    else:
        print('object类型有',index,missing_data[index])
```

```python
data.isnull().sum()
```

```python
#剩余没处理的缺失值
data['reg_preference_for_trad'].value_counts()
```

```python
data['reg_preference_for_trad'].fillna('其他城市',inplace=True)
```

```python
#删除id列
data.drop(['id_name'],axis=1,inplace=True)
```

```python
#众数补充
data['latest_query_time'].fillna(data['latest_query_time'].mode()[0],inplace=True)
data['loans_latest_time'].fillna(data['loans_latest_time'].mode()[0],inplace=True)
```

### 对数值型归一化

```python
for i in data.columns:
    if data[i].dtype!=object:
        data[i]=(data[i]-data[i].min())/(data[i].max()-data[i].min())
data.head()
```

### 对非数值类型One_hot处理
  one-hot的基本思想：将离散型特征的每一种取值都看成一种状态，若你的这一特征中有N个不相同的取值，那么我们就可以将该特征抽象成N种不同的状态，one-hot编码保证了每一个取值只会使得一种状态处于“激活态”，也就是说这N种状态中只有一个状态位值为1，其他状态位都是0。

```python
for i in data.columns:
    if data[i].dtype==object:
        print(i)
```

```python
data.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
data.drop(['bank_card_no'],axis=1,inplace=True)
```

```python
data.reg_preference_for_trad.value_counts()
```

```python
data=data.join(pd.get_dummies(data.reg_preference_for_trad))
data.drop(['reg_preference_for_trad'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.drop(['source'],axis=1,inplace=True)
```

```python
data.drop(['trade_no'],axis=1,inplace=True)
```

```python
#对两个时间的处理
data['time1']=pd.to_datetime(data['latest_query_time'],format='%Y-%m-%d')
data['time2']=pd.to_datetime(data['loans_latest_time'],format='%Y-%m-%d')
```

```python
data['time1'].tail()
```

```python
data['year1']=[i.year for i in data['time1']]
```

```python
data.year1.tail()
```

```python
data['month1']=[i.month for i in data['time1']]
data['day1']=[i.day for i in data['time1']]
data['year2']=[i.year for i in data['time2']]
data['month2']=[i.month for i in data['time2']]
data['day2']=[i.day for i in data['time2']]

```

```python
names=['year1','month1','year2','month2']
for name in names:
    data=data.join(pd.get_dummies(data[name]),lsuffix='_x',rsuffix='_y')
```

```python
data.head()
```

```python
names.append('time1')
names.append('time2')
for name in names:
    data.drop([name],axis=1,inplace=True)
```
