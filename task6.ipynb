### 处理缺失值

```python

missing_data=data.isnull().sum()
for index in missing_data.keys():
    if missing_data[index]==0:
        del missing_data[index]
  

missing_data
```

```python
for index in missing_data.keys():
    if missing_data[index]>2000:
        print('缺失值过多的列',index)
        del missing_data[index]
missing_data
```

```python
data.drop(['student_feature'],axis=1,inplace=True)
```

```python
#对于数值型数据采用平均值补充
for index in missing_data.keys():
    if data[index].dtype!=object:
        data[index].fillna(data[index].mean(),inplace=True)
    else:
        print('object类型有',index,missing_data[index])
```

```python
data.isnull().sum()
```

```python
#剩余没处理的缺失值
data['reg_preference_for_trad'].value_counts()
```

```python
data['reg_preference_for_trad'].fillna('其他城市',inplace=True)
```

```python
#删除id列
data.drop(['id_name'],axis=1,inplace=True)
```

```python
#众数补充
data['latest_query_time'].fillna(data['latest_query_time'].mode()[0],inplace=True)
data['loans_latest_time'].fillna(data['loans_latest_time'].mode()[0],inplace=True)
```

### 对数值型归一化

```python
for i in data.columns:
    if data[i].dtype!=object:
        data[i]=(data[i]-data[i].min())/(data[i].max()-data[i].min())
data.head()
```

### 对非数值类型One_hot处理
  one-hot的基本思想：将离散型特征的每一种取值都看成一种状态，若你的这一特征中有N个不相同的取值，那么我们就可以将该特征抽象成N种不同的状态，one-hot编码保证了每一个取值只会使得一种状态处于“激活态”，也就是说这N种状态中只有一个状态位值为1，其他状态位都是0。

```python
import lightgbm as lgb
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skf=StratifiedKFold(n_splits=5,shuffle=True,random_state=1)
```

```python
params = {
'task': 'train',
'boosting_type': 'gbdt',
'objective': 'binary',
'metric': 'auc',
'num_leaves': 9,
'learning_rate': 0.03,
'feature_fraction_seed': 2,
'feature_fraction': 0.9,
'bagging_fraction': 0.8,
'bagging_freq': 5,
'min_data': 20,
'min_hessian': 1,
'verbose': -1
}
```

```python
layer_train = np.zeros((X.shape[0], 5))
test_pred = np.zeros((X_pred.shape[0],5))
```

```python
for k,(train_index, test_index) in enumerate(skf.split(X, Y)):
	X_train = X[train_index]
	y_train = y[train_index]
	X_test = X[test_index]
	y_test = y[test_index]

	# create dataset for lightgbm
	lgb_train = lgb.Dataset(X_train, y_train)
	lgb_eval = lgb.Dataset(X_test, y_test, reference=lgb_train)

	# train
	gbm = lgb.train(params,
				lgb_train,
				num_boost_round=10000,
				valid_sets=lgb_eval,
				early_stopping_rounds=200)

	pred_y = gbm.predict(X_test, num_iteration=gbm.best_iteration)
	layer_train[test_index, 1] = pred_y

	pred = gbm.predict(X_pred, num_iteration=gbm.best_iteration)
	test_pred[:, k] = pred
```

```python
layer_test = test_pred.mean(axis=1)


X = pd.DataFrame(X)
X['pre']=layer_train[:,1]
X_pred = pd.DataFrame(X_pred)
X_pred['pre'] = layer_test

X = X.values
X_pred = X_pred.values

for cv in range(5):
	X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=cv)
```
